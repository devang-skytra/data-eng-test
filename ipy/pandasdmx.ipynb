```python
from pandasdmx import Request

# Define the source
ecb = Request('ECB')
```

```python
# Retrieve the data (we start at 2016, because are requesting a larger dataset (including other frequencies))
data_response = ecb.data(resource_id = 'EXR', key={'CURRENCY': ['EUR', 'USD']}, params = {'startPeriod': '2020'})
data = data_response.data[0]

# The data will be a pandaSDMC 'DataSet'
type(data)
```
